```python
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
corr_matrix = corr_matrix.fillna(0)
sns.heatmap(corr_matrix, annot=True, annot_kws={"size":12, "color": "black"})
plt.rcParams['figure.figsize'] = (10, 5)
plt.show(block=True)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", annot_kws={"size":10, "color": "black"}, cmap="coolwarm", linewidths=0.5)
plt.tight_layout()
plt.show()
```
